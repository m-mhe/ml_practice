# Multioutput Classification

```python
from sklearn.datasets import fetch_openml
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import matplotlib.pyplot as plt
from xgboost import XGBClassifier
```

```python
def display(img):
    img = img.reshape(28, 28)
    plt.imshow(img, cmap="binary")
    plt.axis("off")
    plt.show()
def genRandomIndex(upbound=60000):
    return np.random.randint(0,upbound, dtype=int)
```

```python
mnistDataset = fetch_openml("mnist_784", as_frame=False)
xTrain, yTrain, xTest, yTest = mnistDataset.data[:60000], mnistDataset.target[:60000], mnistDataset.data[60000:], mnistDataset.target[60000:]
xTrain, yTrain, xTest, yTest
```

```python
display(xTrain[genRandomIndex()])
```

```python
noise = np.random.randint(0, 100, (xTrain.shape))
xTrainNoisy = noise + xTrain
noise = np.random.randint(0, 100, (xTest.shape))
xTestnNoisy = noise + xTest
```

```python
display(xTrainNoisy[genRandomIndex()])
display(xTestnNoisy[69])
```

```python

knnModel = KNeighborsClassifier(
    n_neighbors=3,
    weights="distance",
    metric="minkowski",
    p=2,
    algorithm="auto",
)
```

```python
knnModel.fit(xTrainNoisy, xTrain)
```

```python
randomIndex = genRandomIndex(upbound=10000)
display(xTest[randomIndex])
display(xTestnNoisy[randomIndex])
prediction = knnModel.predict([xTestnNoisy[randomIndex]])
display(prediction)
print(f"Digit: {yTest[randomIndex]}")
```
