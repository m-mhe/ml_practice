# Silent Hill 2 Remake Performance Analysis

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
rawDF = pd.read_csv("SHProto-Win64-Shipping_2026-05-20_23-30-42.csv")
display(rawDF.sample(10))
display(rawDF.describe())
rawDF.info()
```

```python
plt.figure(figsize=(20,7))
plt.plot(np.arange(0, 741), rawDF["fps"], alpha=0.8,)
```

```python
sns.kdeplot(data=rawDF, x="gpu_core_clock")
```

```python
plt.figure(figsize=(20,10))
plt.plot(np.arange(0, 741), rawDF["gpu_core_clock"])
```

```python
sns.kdeplot(data=rawDF, x= "gpu_load")
```

```python
plt.figure(figsize=(12,7))
sns.scatterplot(x= rawDF["gpu_power"], y=rawDF["gpu_load"], hue=rawDF["gpu_core_clock"])
```

```python
plt.plot(rawDF.sort_values("gpu_power")["gpu_power"],rawDF.sort_values("gpu_power")["gpu_core_clock"])
```

```python
plt.figure(figsize=(10,7))
plt.scatter(x=rawDF["gpu_power"], y=rawDF["gpu_core_clock"], alpha=0.5)
```
