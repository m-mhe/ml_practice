# KMEANS VS DBSCAN

```python
from sklearn.cluster import KMeans, DBSCAN
from sklearn.datasets import make_moons
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
X, y = make_moons(n_samples=25_000, noise=0.08, random_state=67)
scaler = StandardScaler()
X = scaler.fit_transform(X)
plt.scatter(X[:,0], X[:,1], s=0.07)
```

```python
kMeansCluster = KMeans(n_clusters=2, max_iter=5000, random_state=67)
dBscanCluster = DBSCAN(eps=0.1, min_samples=3)
```

```python
kMeansCluster.fit(X)
dBscanCluster.fit(X)
```

```python
kMLabels = kMeansCluster.labels_
dBLabels = dBscanCluster.labels_
```

```python
sns.scatterplot(x=X[:,0], y=X[:,1], hue=kMLabels, s=1.7)
plt.title("KMeans")
plt.show()
```

```python
sns.scatterplot(x=X[:,0], y=X[:,1], hue=dBLabels, s=1.7, palette="tab10")
plt.title("DBSCAN")
plt.show()
```
